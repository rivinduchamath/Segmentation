```python
df5 = pd.read_csv("orderIDandITEM.csv", sep=',', index_col=False, dtype='unicode') 
# SELECT order_id, item_id , i.title from reservations r left join items i  On i.id  = r.item_id
```

```python
X = df5.iloc[:,[0,1]].values
```

```python
print(X)
```

```python
wcss = []
for i in range(1,11):
 kmeans = KMeans(n_clusters=i, init='k-means++',random_state=42)
 kmeans.fit(X)
 wcss.append(kmeans.inertia_)
```

```python
sns.set()
plt.plot(range(1,11),wcss)
plt.title('graph')
plt.xlabel('Number')
plt.xlabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters=5, init='k-means++',random_state=0)
Y = kmeans.fit_predict(X)
print(Y)
```

```python
plt.figure(figsize=(8,8))
plt.scatter(X[Y==0,0],X[Y==0,1],s=50,c='green', label='Cluster 1')
plt.scatter(X[Y==1,0],X[Y==1,1],s=50,c='red', label='Cluster 2')
plt.scatter(X[Y==2,0],X[Y==2,1],s=50,c='yellow', label='Cluster 3')
plt.scatter(X[Y==3,0],X[Y==3,1],s=50,c='violet', label='Cluster 4')
plt.scatter(X[Y==4,0],X[Y==4,1],s=50,c='blue', label='Cluster 5')

plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1],s=100, c="cyan", label ='Cenroids')

plt.title('Order Items')
plt.xlabel('Number')
plt.xlabel('WCSS')
plt.show()
```
